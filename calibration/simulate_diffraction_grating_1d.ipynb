```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
b = (1/500)*1e-3  # m/groove 
wvls = np.arange(400, 700, 1)*1e-9

I0 = 1
theta = np.deg2rad( np.arange(-25, 25, 0.1) )
dim = 30e-3 # m
#N = dim/b
N = 10

sigma = (2*np.pi/wvls[:,np.newaxis]) * b * np.sin(theta[np.newaxis,:])
```

```python
s = np.sin(sigma/2)
sN = np.sin(N*sigma/2)
```

```python
I = I0*(s**2)*(sN**2)/(((sigma/2)**2) * (s**2))
```

```python
plt.figure()
plt.plot(np.rad2deg(theta), I.T)
plt.ylim([0,1])

plt.figure()
plt.plot(np.rad2deg(theta), I[200,:])
plt.ylim([0,1])

plt.figure()
plt.plot(np.rad2deg(theta), I.T)

plt.figure()
plt.plot(np.rad2deg(theta), I[200,:])
```
