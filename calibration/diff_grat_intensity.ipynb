```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

### 450nm

```python
pixel_half_w = 2
```

#### m = 0 order

```python
m0_img = cv2.imread('./experiments/test_2023_03_28_14_48/450nm_50ms/capture_0000.png')
m0_img = cv2.cvtColor(m0_img, cv2.COLOR_BGR2RGB)
```

```python
m0_img.shape
```

```python
plt.imshow(m0_img)
```

```python
top_x, top_y, bottom_x, bottom_y = 131,221,123,559
roi_m0_top = m0_img[top_y-pixel_half_w: top_y+pixel_half_w ,top_x-pixel_half_w: top_x+pixel_half_w]
roi_m0_bottom = m0_img[bottom_y-pixel_half_w: bottom_y+pixel_half_w ,bottom_x-pixel_half_w: bottom_x+pixel_half_w]
```

```python
print(roi_m0_top.reshape(-1,3).mean(axis = 0))
print(roi_m0_bottom.reshape(-1,3).mean(axis = 0))
```
