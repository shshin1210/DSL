```python
import numpy as np
import cv2
import os
import matplotlib.pyplot as plt
```

```python
# different exposure time
exposure_time = np.array([160, 180, 200, 220, 240, 260, 280, 300])

# center points for all patches
all_patches_center_pts = np.array([[161, 137], [287,147], [407,151], [528, 153], [647,161], [767,164],
[150,263], [280,265], [400,260], [532, 272], [650,270], [770,280],
[145,375], [278,385], [407,387], [534,394], [657,390],  [774,397],
[141,500], [273,507], [403,514], [530, 513], [657,513], [782, 513]])
```

```python
rgb_list = []

for i in range(len(exposure_time)):
    fn = '%sms_crop/calibration00/capture_%04d.png'%(exposure_time[i], 0)
    image = cv2.imread(fn, -1)[:,:,::-1] / 65535.
    
    rgb_list.append(image[all_patches_center_pts[:,1],all_patches_center_pts[:,0]].flatten().mean())   
    
```

```python
plt.figure()
plt.xlabel('exposure_time')
plt.ylabel('intensity')
plt.xticks(np.arange(0, 8), exposure_time)
plt.plot(rgb_list)
```
