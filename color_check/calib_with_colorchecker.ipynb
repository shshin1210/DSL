```python
im = cv2.imread('./dataset/colorchecker_captured.png', -1)/65535.
im = im[:,:,::-1]
plt.figure(1)
plt.imshow(im)

x, y, xstep, ystep, psize = 230, 250, 135, 125, 50

C = 6
R = 4

rgb_meas = []
for r in range(R):
    for c in range(C):
        cc = x + c*xstep
        rr = y + r*ystep
        plt.figure(1)
        plt.plot(cc,rr, '*')

        patch = im[rr-psize//2:rr+psize//2,cc-psize//2:cc+psize//2,:]
        plt.figure(figsize=(2,2))
        plt.imshow(patch)
        plt.title('patch (%d,%d)'%(r,c))

        rgb_meas.append( patch.mean(axis=(0,1)) )

rgb_meas = np.array(rgb_meas)

plt.imshow(rgb_meas[None,:,:])
plt.title('Captured color patch RGB')
plt.figure()
plt.plot(rgb_meas)
plt.xlabel('patch index')
plt.ylabel('intensity')
plt.legend(['R', 'G', 'B'])
```

```python
rgb_meas1 = np.hstack([rgb_meas, np.ones((N,1))])
print(rgb_meas1.shape, rgb_s.shape)
```

```python
A = rgb_meas1
b = rgb_s


x, residuals, rank, s = np.linalg.lstsq(A, b, rcond='warn')
print(x, residuals, rank)

meas2syn_mat = x

def meas2syn(im_meas, meas2syn_mat):
    R,C,_ = im_meas.shape
    im_meas1 = np.hstack([im_meas.reshape(R*C,3), np.ones((R*C,1))])
    im_syn = (im_meas1@meas2syn_mat).reshape(R,C,3)
    return im_syn


rgb_meas2syn = meas2syn(rgb_meas[None,:,:], meas2syn_mat)

plt.figure()
plt.imshow(rgb_meas[None,:,:])
plt.title('Captured color patch RGB')

plt.figure()
plt.imshow(rgb_meas2syn)
plt.title('Meas2syn color patch RGB')

# plt.figure()
# plt.imshow(rgb_ss[None,:,:])
# plt.title('Syn color patch RGB')

plt.figure()
plt.imshow(np.abs(rgb_meas2syn-rgb_s[None,:,:]))
plt.title('Diff')
```

```python
def vis(data):
    illum_num = 40
    max_images_per_column = 5
    num_columns = (illum_num + max_images_per_column - 1) // max_images_per_column
    plt.figure(figsize=(10, 3*num_columns))

    for c in range(num_columns):
        start_index = c * max_images_per_column
        end_index = min(start_index + max_images_per_column, illum_num)
        num_images = end_index - start_index
                
        for i in range(num_images):
            plt.subplot(num_columns, num_images, i + c * num_images + 1)
            plt.imshow(data[i + start_index], vmin=0., vmax=1.)
            plt.axis('off')
            plt.title(f"Image {i + start_index}")
                    
            if i + start_index == illum_num - 1:
                plt.colorbar()

    plt.show()
```
