```python
import pandas as pd
import matplotlib.pyplot as plt


# Replace 'file.csv' with the path to your CSV file
file_path = './dataset/MacbethColorChecker.csv'

# Read the CSV file and store it as a DataFrame
df = pd.read_csv(file_path, header=None, skiprows=2)

# Display the DataFrame

wvls_CC = df.values[1:,0]
wvls_CC = [float(x) for x in wvls_CC] 
ref_CC = df.values[1:,1:-2]

ind_min = 6
ind_max = 61
wvls_CC = wvls_CC[ind_min:ind_max:2] 
ref_CC = ref_CC[ind_min:ind_max:2,:]

print(wvls_CC, ref_CC)

plt.figure()
plt.plot(wvls_CC, ref_CC)
plt.title('colorchecker reflectance')
plt.xlabel('wavelength [nm]')
```

```python

# Replace 'file.csv' with the path to your CSV file
file_path = './dataset/camera.csv'

# Read the CSV file and store it as a DataFrame
df = pd.read_csv(file_path, header=None, skiprows=2)

# Display the DataFrame
print(df)

wvls_CRF = df.values[0:,0]
wvls_CRF = [float(x) for x in wvls_CRF] 
CRF = df.values[0:,1:]

print(wvls_CRF)

plt.figure()
plt.plot(wvls_CRF, CRF)
plt.title('CRF')
plt.xlabel('wavelength [nm]')
plt.legend(['R', 'G', 'B'])
```

```python

# Replace 'file.csv' with the path to your CSV file
file_path = './dataset/projector.csv'

# Read the CSV file and store it as a DataFrame
df = pd.read_csv(file_path, header=None, skiprows=2)

# Display the DataFrame
print(df)

wvls_SEF = df.values[0:,0]
wvls_SEF = [float(x) for x in wvls_SEF] 
SEF = df.values[0:,1:]
print(wvls_SEF)
plt.figure()
plt.plot(wvls_SEF, SEF)
plt.title('SEF')
plt.xlabel('wavelength [nm]')
plt.legend(['R', 'G', 'B'])
```

```python
rad = SEF.sum(axis=1)
plt.plot(wvls_SEF, rad)
plt.title('radiance from projector')
plt.xlabel('wavelength [nm]')
```

```python
rad_ref = (rad[:,None]*ref_CC)
plt.plot(wvls_SEF, rad_ref)
plt.title('radiance reflected from CC')
plt.xlabel('wavelength [nm]')
```
