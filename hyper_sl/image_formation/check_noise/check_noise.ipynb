```python
import torch
import matplotlib.pyplot as plt
```

```python
without_noise = torch.load('/home/shshin/Scalable-Hyperspectral-3D-Imaging/dataset/image_formation/check_noise/without_noise.pt')
with_noise_01 = torch.load('/home/shshin/Scalable-Hyperspectral-3D-Imaging/dataset/image_formation/check_noise/with_noise.pt')
with_noise_001 = torch.load('/home/shshin/Scalable-Hyperspectral-3D-Imaging/dataset/image_formation/check_noise/with_noise_0.01.pt')
with_noise_005 = torch.load('/home/shshin/Scalable-Hyperspectral-3D-Imaging/dataset/image_formation/check_noise/with_noise_0.05.pt')
with_gaussian_noise = torch.load('/home/shshin/Scalable-Hyperspectral-3D-Imaging/dataset/image_formation/check_noise/with_gaussian_noise.pt')
with_gaussian_noise_005 = torch.load('/home/shshin/Scalable-Hyperspectral-3D-Imaging/dataset/image_formation/check_noise/with_gaussian_noise_0.05.pt')
```

```python
without_noise = without_noise.reshape(640,640,42,3)
with_noise_01 = with_noise_01.reshape(640,640,42,3)
with_noise_001 = with_noise_001.reshape(640,640,42,3)
with_noise_005 = with_noise_005.reshape(640,640,42,3)
with_gaussian_noise = with_gaussian_noise.detach().cpu().numpy()
with_gaussian_noise = with_gaussian_noise.reshape(640,640,42,3)
with_gaussian_noise_005 = with_gaussian_noise_005.detach().cpu().numpy()
with_gaussian_noise_005 = with_gaussian_noise_005.reshape(640,640,42,3)
```

```python
plt.figure(figsize =(20,5))
plt.subplot(161),plt.imshow(without_noise[:,:,1,0]), plt.colorbar(), plt.title('no noise')
plt.subplot(162),plt.imshow(with_noise_01[:,:,1,0]), plt.colorbar(), plt.title('noise 0.1')
plt.subplot(163),plt.imshow(with_noise_005[:,:,1,0]), plt.colorbar(), plt.title('noise 0.05')
plt.subplot(164),plt.imshow(with_noise_001[:,:,1,0]), plt.colorbar(),plt.title('noise 0.01')
plt.subplot(165),plt.imshow(with_gaussian_noise[:,:,1,0]), plt.colorbar(),plt.title('gaussian noise')
plt.subplot(166),plt.imshow(with_gaussian_noise_005[:,:,1,0]), plt.colorbar(),plt.title('gaussian noise 0.05')
```
